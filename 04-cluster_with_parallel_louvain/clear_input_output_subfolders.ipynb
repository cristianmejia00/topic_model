# Clear `input/` and `output/` Subfolders

This notebook helps you safely clear the contents of the `input/` and `output/` folders under `04-cluster_with_parallel_louvain/`.

Workflow:
1. Run the setup cell.
2. Run the preview cell to see what would be removed.
3. Set `EXECUTE = True` in the final cell and run it to delete contents.

```python
from pathlib import Path
import shutil

def resolve_base_dir() -> Path:
    cwd = Path.cwd()
    if (cwd / 'input').exists() and (cwd / 'output').exists():
        return cwd

    candidate = cwd / '04-cluster_with_parallel_louvain'
    if (candidate / 'input').exists() and (candidate / 'output').exists():
        return candidate

    raise FileNotFoundError(
        'Could not locate 04-cluster_with_parallel_louvain with input/ and output/ folders. '
        'Run this notebook from the project root or from 04-cluster_with_parallel_louvain/.'
    )

BASE_DIR = resolve_base_dir()
INPUT_DIR = BASE_DIR / 'input'
OUTPUT_DIR = BASE_DIR / 'output'

print('Base directory:', BASE_DIR)
print('Input directory:', INPUT_DIR)
print('Output directory:', OUTPUT_DIR)
```

```python
def folder_snapshot(folder: Path, max_items: int = 50):
    entries = sorted(folder.iterdir(), key=lambda p: p.name.lower())
    files = sum(1 for p in folder.rglob('*') if p.is_file())
    dirs = sum(1 for p in folder.rglob('*') if p.is_dir())

    print(f'\n[{folder.name}]')
    print(' Top-level entries:', len(entries))
    print(' Nested files:', files)
    print(' Nested directories:', dirs)

    for p in entries[:max_items]:
        kind = 'DIR ' if p.is_dir() else 'FILE'
        print(f'  {kind}  {p.name}')

    if len(entries) > max_items:
        print(f'  ... and {len(entries) - max_items} more')

folder_snapshot(INPUT_DIR)
folder_snapshot(OUTPUT_DIR)
```

```python
def clear_folder_contents(folder: Path, execute: bool = False):
    entries = sorted(folder.iterdir(), key=lambda p: p.name.lower())
    action = 'DELETE' if execute else 'DRY-RUN'
    print(f'\n[{action}] {folder}')

    if not entries:
        print('  Nothing to remove.')
        return

    for p in entries:
        print(f'  - {p.name}')
        if execute:
            if p.is_dir():
                shutil.rmtree(p)
            else:
                p.unlink()

    if execute:
        print('  Done.')

# Safety switch: keep False first, review output, then set True to actually delete.
EXECUTE = False

# Toggle these if you want to clear only one folder.
CLEAR_INPUT = True
CLEAR_OUTPUT = True

if CLEAR_INPUT:
    clear_folder_contents(INPUT_DIR, execute=EXECUTE)
if CLEAR_OUTPUT:
    clear_folder_contents(OUTPUT_DIR, execute=EXECUTE)

print('\nFinished. EXECUTE =', EXECUTE)
```

```python
# Optional: quick post-check
folder_snapshot(INPUT_DIR)
folder_snapshot(OUTPUT_DIR)
```
